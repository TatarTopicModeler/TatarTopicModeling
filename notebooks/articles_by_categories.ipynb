```python
from bs4 import BeautifulSoup
import requests

page = 'https://tt.wikipedia.org/wiki/%D0%A2%D3%A9%D1%80%D0%BA%D0%B5%D0%BC:%D0%A2%D0%B0%D1%82%D0%B0%D1%80_%D1%80%D0%B8%D0%B7%D1%8B%D0%BA%D0%BB%D0%B0%D1%80%D1%8B'
cont = requests.get(page).content

soup = BeautifulSoup(cont)
articles_block = soup.find('div', lang='tt', dir='ltr', class_='mw-content-ltr')
articles_block.find_all('a', class_=None)
```

```python
import os
from pathlib import Path
import pandas as pd

PATH = Path(os.getcwd()).parent
df = pd.read_csv(PATH / 'data/raw/cats.csv')
df.sort_values('count', ascending=False, inplace=True)

df['title'] = df['title'].str.lower()
stop = ['торак пункт', 'елга', 'фамилия', 'википедия', 'wikidata', 'туу', 'үлем', 'казан', 'уфа', 'вафат', 'сурәтләр', 'саннар', 'авыл']
for s in stop:
    df = df[ ~df['title'].str.contains(s) ]

i = 0
for index, row in df.iterrows():
    print(row['title'], '\t', row['count'])
    i += 1
    if i > 100:
        break
```

```python
import nltk
from collections import Counter

words = [x  for doc in df['title'].values.astype('U') for x in nltk.word_tokenize(doc)]
cnt = Counter(words)
cnt.most_common(100)
```

```python
topics = ['Астрономия', 'Биология', 'Химия', 'Физика', 'Математика', 'География', 'тарих', 
          'әдәбият', 'Фәлсәфә', 'Психология', 'Сәясәт', 'Икътисад', 'хокук']
topics = [t.lower() for t in topics]
total_art = 0

for t in topics:
    cat = df[df['title'].str.contains(t)]
    cnt = cat["count"].sum()
    total_art += cnt
    print(f'{t}, \t cats: {len(cat)}, \t articles: {cnt}')

print()
print('Total articles:', total_art)
cat
```
