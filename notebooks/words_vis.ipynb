```python
import os
import pandas as pd
from pathlib import Path

PATH = Path(os.getcwd()).parent
df = pd.read_csv(PATH / 'data/processed/contents.csv')
df
```

```python
df.dropna(inplace=True)
df = df[~df['title'].str.contains('Калып:')]  # remove templates
```

## LDA

```python
import nltk
import pyLDAvis.gensim

from gensim.corpora import Dictionary
from gensim.models import LdaModel, HdpModel, CoherenceModel

TOPICS = 13
SEED = 5

pyLDAvis.enable_notebook()

documents = df['preproc'].values.astype('U')
train_texts = [nltk.word_tokenize(d) for d in documents]
dictionary = Dictionary(train_texts)
corpus = [dictionary.doc2bow(text) for text in train_texts]

ldamodel = LdaModel(corpus=corpus, num_topics=TOPICS, id2word=dictionary, random_state=SEED)
pyLDAvis.gensim.prepare(ldamodel, corpus, dictionary)
```

```python
ldatopics = ldamodel.show_topics(formatted=False)
ldatopics
```

## HDP

```python
hdpmodel = HdpModel(corpus=corpus, id2word=dictionary, random_state=SEED)

pyLDAvis.gensim.prepare(hdpmodel, corpus, dictionary, mds='tsne')
```

```python
print(len(hdpmodel.show_topics()))
hdpmodel.show_topics()
```
